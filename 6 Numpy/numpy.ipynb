## Numpy vs List

1. NumPy arrays are faster (optimized in C)
2. NumPy stores data in continuous memory blocks
3. Uses less memory (optimized storage)
4. Built-in mathemetical functions
5. Supports direct vectorized operations

```python
import time

# list
size = 1_000_000
py_list = list(range(size))
start_time = time.time()
sq_list = [x ** 2 for x in py_list]
end_time = time.time()
print(f"List calculation time = {end_time - start_time}")

# numpy

size = 1_000_000
np_array = np.array(py_list)
start = time.time()
sq_list = np_array ** 2 
end = time.time()
print(f"numpy calculation time = {end - start}")
```

```python
import sys

print(f"memory occupied by python list = {sys.getsizeof(py_list) *len(py_list)} bytes")
print(f"memory occupied by numpy = {np_array.nbytes} bytes")
```

```python
arr = np.array([1, 2, 3, 5, 4, 3])

print(arr[arr>2]) # all ele greater than 2
print(arr[arr % 2 ==0]) # we get all even values
```

## slicing

```python
arr = np.array([1, 2, 3, 4, 5, 6])
print(arr[::]) # get all element
print(arr[1:4]) # idx 1 is included while idx 4 is excluded
print(arr[1:])
print(arr[::2]) # last index is step
```

## Copy and view
- In list slicing create a new copy in another memory location
- numpy creates only view

```python
# list
print("\n list\n")
nums = ([1, 2, 3, 4, 5, 6])
sub_list = nums[1:3]  # new memory location
print("sublist after slicing :", sub_list)

sub_list[0] = 200
print("updated sublist :", sub_list)
print("original :", nums) # oiginal remains same


print("\n numpy\n")
# numpy
num = np.array([1, 2, 3, 4, 5, 6])
sub_lists = num[1:3]
print("sublist after slicing :",sub_lists)

sub_lists[0] = 200
print("updated sublist :", sub_lists)
print("original :",num)
```
